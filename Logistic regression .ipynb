```python
import pandas as pd

url=r'\\192.168.0.118\d\MachineLearning\Data\Credit.xls'
crdt=pd.read_excel(url,sheet_name='Data')
crdt.head()
```

```python
crdt.shape
```

```python
crdt.describe().T
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics 

crdt.columns = crdt.columns.str.strip().str.replace('#','')
```

```python
crdt.RESPONSE.value_counts()
```

```python
cedt=crdt.drop(['OBS'],axis =1)
```

```python
cedt.corr()
```

```python
x = cedt[list(cedt.columns)[:-1]]
y = cedt['RESPONSE']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y,random_state=0)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
from sklearn.linear_model.logistic import LogisticRegression

model= LogisticRegression(solver='liblinear',max_iter=100)
l1reg=model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print("accuracy :\n", metrics.accuracy_score(y_test,y_pred))
print("classification:\n", metrics.classification_report(y_test,y_pred))
print("confusion matrix:\n", metrics.confusion_matrix(y_test,y_pred))
```
